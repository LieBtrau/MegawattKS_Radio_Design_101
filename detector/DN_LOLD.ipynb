# Design Note Locked Oscillator Limiter Discriminator

## Purpose of the design
Takes FM-modulated input at 10.7MHz and demodulates it into an audio output.

## Details
The original 12V design from Elektor has been simulated.  In the simulation, the ceramic filter has been replaced by a series LC-circuit. 

### Limiter
As the circuit is drawn, it's more of a programmable gain amplifier than a limiter.  For a limiter, both diodes would need to be reverse biased.
In this circuit, the diodes are more or less forward biased.  The higher the voltage over R37, the more forward biased they become and the more the output signal is reduced.

In the original circuit, there's a 10n and 1K resistor from the cathode of D1 to ground.  I see no reason for this.  It adds an extra DC-offset, but the output is AC-coupled anyway.

The limiter in the original schematic is dimensioned as such that the maximum voltage over R37 is 0.6V.  Because the supply voltage is only 5 V in our case, R39 has been reduced from 18K to 6.2K.

The "limiter" seems to deliver the broadest range when the voltage over C31 is regulated from 0 to 1.4V.  To limit the amount of current through the potentiometer, we can use a transistor to buffer it.  We can now use a 10K potentiometer, which will consume less current than a 1K potentiometer.  R37 is added in series to limit the voltage on the anode of Q11 to about 2V.


```python
# Current through R37

Vcc = 5
R39 = 6.2e3
I_R37 = Vcc / (R39 + 1e3 + 1e3)
print(f"Current through R37: {I_R37*1e3:.2f} mA")
```




### 12V bias settings
* Q9 (left transistor) : Ic = 1.5mA, Vce = 10V
* Q10 (right transistor) : Ic = 300µA, Vce = 10V

### 5V bias settings
The circuit was then modified to operate at 5V.
* Q9 (left transistor) : Ic = 1.5mA, Vce = 3.4V
* Q10 (right transistor) : Ic = 100µA, Vce = 3V

R35 has been increased from 18k (300µA Ic10) to 47K (220µA Ic10) because it yields a more linear frequency response.

### De-emphasis
R36/C29 in the design have been changed according to the de-emphasis network in the NJM14570 datasheet.
