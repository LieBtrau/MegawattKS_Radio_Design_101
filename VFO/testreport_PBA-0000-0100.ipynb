# Test result details
## [TIR1](#tir1) fixing lower end of the tuning range
The disadvantage of this kind of tuning is that it also impacts the gain of the oscillator.  At F_TUNE=1 V, the oscillation peak-peak voltage is only 35 mV, while at F_TUNE=4 V, it's 88 mV.

Without C2, at F_TUNE=4 V, the oscillation frequency is 119.5 MHz.  We would like this to be around 108 MHz-10.7MHz=97.3 MHz

```python
import math
f1 = 97.3e6
ω1 = 2 * math.pi * f1
L = 100e-9
C1 = 1 / (math.pow(ω1,2) * L)
print(f"C1 = {C1*1e12:.0f} pF")
```

With a 100nH inductor, there's 18pF parallel capacitance at 119.5MHz.  To get the frequency down to 97.3 MHz, we need to add 9pF in parallel.

### With C2 = 8.2 pF
* F_TUNE = 1 V: 74.3MHz, 37mVpp on R9 (no further components placed on the board).  Ok because it's smaller than 77.3 MHz.
* F_TUNE = 4 V: 98.1 MHz, 75 mVpp on R9 (no further components placed on the board)

## [TIR4](#tir4) fixing higher end of the tuning range
The upper frequency can't be reached.  Either we have to reduce the load capacitance or reduce the inductance.  If we reduce the inductance, we'll need a larger capacitance variation to cover the whole FM-band.  Moreover there's no 68nH readily available.

```python
import math
f1 = 100e6
ω1 = 2 * math.pi * f1
L = 82e-9
C1 = 1 / (math.pow(ω1,2) * L)
print(f"C1 = {C1*1e12:.0f} pF")
```

With the 82 nH inductor, we have 36 pF parallel with it in the circuit.  To get to 100 MHz, we need to reduce this capacitance to 31pF.

```python
C12 = 39e-12
C11 = 27e-12
C_D4 = 12e-12 # with F_TUNE = 4 Vdc

C_tot = 1 / (1/C11 + 1/C12) + C_D4
print(f"C_tot = {C_tot*1e12:.0f} pF")
```

So it seems that we have 8pF of stray capacitance in the circuit.

```python
C12 = 22e-12
C11 = 22e-12
C_D4 = 12e-12 # with F_TUNE = 4 Vdc
C_stray = 8e-12 # estimated

C_tot = 1 / (1/C11 + 1/C12) + C_D4 + C_stray
print(f"C_tot = {C_tot*1e12:.0f} pF")
```
