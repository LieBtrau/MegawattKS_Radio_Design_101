# Design note for the FM-front end
## Background
Almost all historical discrete radios use a common-base as first stage.  Only the grundig satellite 1400 uses a JFET.  As I have already designed a common-base stage, we can reuse that.  Its input 1dB compression point (Pi1dB) is around -25 dBm (36 mVpp in 50 ohm).  This isn't great, but I don't have very strong stations around here, so it should be sufficient.

The MegawattKS youtube series focuses on the design of a 50ohm input LNA with the final goal of converting it to a Q-enhanced LNA.  The author is using a discone antenna with a 50ohm impedance.  In a portable radio, we'll have to use a whip antenna, whose impedance is far from 50ohm.

The purpose of this design note is to design a matching network for a whip antenna.
