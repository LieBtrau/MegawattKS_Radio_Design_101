# Mixer
Some of the resources use a single JFET mixer, but that has strong dependence on device to device variations, and it leaks the LO signal into the output quite strongly.  An alternative would be to use a diode ring mixer, but that's an expensive solution (ADE1+ is > $6).
Because of the high amplitude of the signals involved, the mixer is implemented as a switching mixer.  See EMRFD fig. 5.10 for an example.  This switching mixer topology has about 6dB of loss.

A diplexer is use to separate the RF port from the IF port.  The crossover frequency for the mixer is the geometrical mean of the RF and IF frequencies.  On fig 5.10, a first order diplexer is shown, but on EMRFD p. 3.37 third order multiplexers are shown.  Even higher order diplexers are possible, see ARRL 2018 handbook, §10.10.5 Diplexer filters.

# Remarks after testing
A switching mixer might be cheaper, but it has lots of harmonics.  A comparison simulation with a diode ring mixer can be found [here](../IF1_mixer/simulation/ringmod_vs_switcher.asc).  A diode ring mixer effectively suppresses the LO and incoming RF-signals, which the switching mixer does not.  A diplexer can remedy the leakage, but that adds complexity and cost as well.

The ADE-1+ would be a suitable diode ring mixer.  IF output is down to DC.

A simple switching mixer works in this application, because the LO and RF signals are quite far from the IF frequency.  The diplexer can easily suppress the LO and RF signals, so that the IF output is clean enough. 

```python
import math

f_LF = 200e3 # Low-frequency cutoff (200 kHz)
f_IF = 10.7e6 # Intermediate frequency (10.7 MHz)
f_co = math.sqrt(f_LF * f_IF) # Cross-over frequency for the bandpass filter
print(f"Calculated cross-over frequency for the bandpass filter: {f_co/1e3:.2f} kHz")
```

```python
# Third order Low Pass High Pass Diplexer, EMRFD p3.38
import math
f = f_co
ω = 2 * math.pi * f

def calc_L(ω, Z):
    """Calculate inductance for a given ω and impedance."""
    return Z / ω

def calc_C(ω, Z):
    """Calculate capacitance for a given ω and impedance."""
    return 1 / (ω * Z)

L_25 = calc_L(ω, 25)  # L for 25 ohm
C_25 = calc_C(ω, 25)  # C for 25 ohm
L_375 = calc_L(ω, 37.5)  # L for 37.5 ohm
C_375 = calc_C(ω, 37.5)  # C for 37.5 ohm
L_75 = calc_L(ω, 75)  # L for 75 ohm
C_75 = calc_C(ω, 75)  # C for 75 ohm

print(f"(25 ohm): L = {L_25*1e6:.2f} µH, C = {C_25*1e9:.2f} nF")
print(f"(37.5 ohm): L = {L_375*1e6:.2f} µH, C = {C_375*1e9:.2f} nF")
print(f"(75 ohm): L = {L_75*1e6:.2f} µH, C = {C_75*1e9:.2f} nF")
```

Actual values for the diplexer are rounded off to E24 values for the capacitors and E12 values for the inductors.  The capacitors will be 0603, while the inductors will be 1008.
LTSpice simulations (switching_mixer.asc) show that these diplexer values are sufficient to achieve the desired isolation between the RF and IF ports.
