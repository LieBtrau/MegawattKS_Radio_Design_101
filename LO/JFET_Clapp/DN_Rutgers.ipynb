The calculations have been redefined in a more general way.

```python
import math 
f0 = 10.5e6 # Oscillation frequency in Hz
C = 500 / (f0 * 1e6) # Capacitance in F
Ca = 4 * C # Capacitance in F

# Calculate the effective capacitance C0 and the inductance L for the Clapp oscillator
C0 = 1 / ((1/C) + (2/Ca)) # Capacitance in F
L = 1 / ((2 * math.pi * f0)**2 * C0) # Inductance in H
Z_choke = 1500 # Impedance of the choke in Ohm
L_choke = Z_choke / (2 * math.pi * f0) # Inductance of the choke in H

print(f"C = {C*1e12:.1f} pF")
print(f"Ca = {Ca*1e12:.1f} pF")
print(f"L = {L*1e6:.2f} µH")
print(f"L_choke = {L_choke*1e6:.2f} µH")
```

```python
# Comparison of the calculated values with the ones from SSDRA

Z_Ca = 1 / (2 * math.pi * f0 * Ca) # Impedance of Ca in Ohm
Z_L = 2 * math.pi * f0 * L # Impedance of L in Ohm
Z_L_choke = 2 * math.pi * f0 * L_choke # Impedance of the choke in Ohm

print(f"Z_Ca = {Z_Ca:.1f} Ohm")
print(f"Z_L = {Z_L:.1f} Ohm")
print(f"Z_L_choke = {Z_L_choke:.1f} Ohm")
```

Compare that to the SSDRA values:

* Z_Ca = 45 ohms
* Z_L = 260 ohms
* Z_L_Choke = 4500 ohms

So the SSDRA design uses a lower inductance and higher capacitance than the Rutgers design.  The RSGB fig 4.14 values are close to the SSDRA values.

These values have been adjusted to standard component values.

## N-JFET
The J309 works as well as the J310, but consumes less current.  By adding drain resistors, the current in the JFETs is further reduced.  The total power in JFET + drain resistor is lower than in a JFET without drain resistor.

## AGC elements
The gate of the JFET is connected with a diode to a DC-source.  This can be a zener diode.  A zener voltage around 3V is probably a good starting point.  A SOD-523 zener diode has been chosen because it fits well with 0603 in the design.  The zener current in the datasheet is between 1 mA and 5 mA.  The 1 K resistor sets a 2 mA zener current for a 5 V power supply.

### Power supply variation
* 12 V power supply: D2 should be around 8.2 V in that case.
* 5 V power supply: replace D2 by a 0 ohm resistor and remove the other AGC elements.

## Frequency adjustment
L2 and C3 have been swapped with respect to the original design.  C3 can be replaced by a  parallel combination of a fixed capacitor and a trim capacitor.  Do not connect capacitors in series, because the ESRs will add up and lower the quality factor, although this is probably insignificant.

## Buffer stage
The output level of the buffer stage can be adjusted by changing the value of C5.  The higher the value, the higher the output level, but that will also increase the loading of the tank circuit.  

A resistor has been added in the drain of the JFET to reduce the power consumption in the JFET.  Without it, the JFET would consume around 150 mW, which is close to the 225 mW maximum.  Now it runs only at 55 mW and the drain resistor runs at 10 mW.

## RF choke
A choke with a high self resonant frequency should be used.  The series resistance is not that critical, but it should be low enough to not disrupt the JFET's DC-bias.  The current rating should be high enough to not cause saturation.  In a 3225-package, 22 µH and 47 µH inductors are available with SRF well into the tens of MHz.  The series resistance is below 1 ohm.

## Shielding
The transistor can be included inside the shielding enclosure because its power consumption is around 175 µW, so it won't add to the heating.
